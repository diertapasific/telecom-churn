#Random Forest

```python
#dataframes for our model
X = telecom_df.drop(['class', 'area_code', 'phone_number'], axis = 'columns')
y = telecom_df['class']
```

```python
X.shape
```

```python
y.shape
```

```python
#split train and test datas
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 150)
```

```python
#train model using Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train.values.ravel())
```

```python
#feature importances per column in X
rf.feature_importances_
```

```python
#plot feature importance

feat_score = pd.DataFrame({"Fraction of variables affected" : rf.feature_importances_}, index = X.columns)
feat_score = feat_score.sort_values(by = "Fraction of variables affected")
feat_score.plot(kind = 'barh', figsize = (10, 5))
sns.despine()
```
